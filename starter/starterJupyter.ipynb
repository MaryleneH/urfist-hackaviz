```python
# -------------------------------------------- #
# A - Data load
# -------------------------------------------- #

import unicodecsv
import pandas as pd
from IPython.display import display, HTML
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# function read csv
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)

#load CSV as list
filename = 'Version simplifié - Viz For Social Good 360 impact data.csv'
alist = read_csv(filename)

len(alist)
```

```python
df = pd.DataFrame(alist)
df.head()
```

```python
# -------------------------------------------- #
# B - Data parsing
# -------------------------------------------- #

df.replace('', np.nan, inplace = True)
#df['female_unpaid_domestic_work'] = df['female_unpaid_domestic_work'].astype(float)
```

```python
attribute_to_float = ['female_unpaid_domestic_work', 'male_unpaid_domestic_work', 'Wife_decision_health',
                  'Wife_decision_household_purchase', 'Women_in_food_decision', 'death_injury_15_34_female',
                  'death_injury_15_34_male', 'death_injury_35_59_female', 'death_injury_35_59_male',
                  'female_mortality_poisonning', 'male_mortality_poisonning', 'male_mortality_poisonning',
                  'female_mortality_air_pollution', 'male_mortality_air_pollution',
                  'cooking_inside', 'cooking_outdoors', 'fuel_crops', 'fuel_charcoal', 'fuel_dung', 'fuel_electricity',
                  'fuel_gas', 'fuel_straw', 'fuel_wood', 'clean_fuel']

for attribute in attribute_to_float :
    df[attribute] = df[attribute].astype(float)
```
